```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("./ML471_S1_Datafile_Practice.csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

# First-order differencing
df["Differenced_Close"] = df["Close"].diff()

# Plot
plt.figure(figsize=(18, 5))
plt.plot(df["Date"], df["Differenced_Close"], label="Differenced Close Price")
plt.title("Differenced Close Price")
plt.xlabel("Date")
plt.ylabel("Differenced Value")
plt.legend()
plt.show()
```
