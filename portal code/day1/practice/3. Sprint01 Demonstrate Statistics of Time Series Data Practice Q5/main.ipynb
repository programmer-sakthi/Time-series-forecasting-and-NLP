```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Load data
df = pd.read_csv("./ML471_S1_Datafile_Practice.csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

# Simple Exponential Smoothing
ses = SimpleExpSmoothing(df["Close"], initialization_method="estimated")
ses_fit = ses.fit(smoothing_level=0.2, optimized=False)

# Plot
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Original Close")
plt.plot(df["Date"], ses_fit.fittedvalues, label="SES Smoothed")
plt.title("Simple Exponential Smoothing of Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```

```python
# Moving Averages
df["SMA_10"] = df["Close"].rolling(window=10).mean()
df["SMA_30"] = df["Close"].rolling(window=30).mean()

# Plot
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Close Price")
plt.plot(df["Date"], df["SMA_10"], label="10-Day SMA")
plt.plot(df["Date"], df["SMA_30"], label="30-Day SMA")
plt.title("Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```
