```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset (update file name if needed)
df = pd.read_csv('./ML471_S1_Datafile_Concept.csv')

# Convert Date column to datetime
df["DATE"] = pd.to_datetime(df["DATE"])

# Sort by date
df = df.sort_values("DATE")

```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["DATE"], df["Consumption"], label="Electricity Consumption")
plt.xlabel("DATE")
plt.ylabel("Consumption")
plt.title("Electricity Consumption Over Time")
plt.legend()
plt.show()
```

```python
# Monthly average consumption
monthly_avg = df.resample("ME", on="DATE")["Consumption"].mean()

plt.figure(figsize=(12, 5))
plt.plot(monthly_avg.index, monthly_avg.values)
plt.xlabel("Month")
plt.ylabel("Average Consumption")
plt.title("Monthly Average Electricity Consumption")
plt.show()
```
