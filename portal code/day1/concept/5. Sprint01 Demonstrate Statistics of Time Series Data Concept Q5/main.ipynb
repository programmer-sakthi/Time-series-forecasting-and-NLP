```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("./ML471_S1_Datafile_Concept.csv")
```

```python
df["DATE"] = pd.to_datetime(df["DATE"])

# Sort by date
df = df.sort_values("DATE")

# First-order differencing
df["Diff_Consumption"] = df["Consumption"].diff()

# Plot differenced series
plt.figure(figsize=(12, 6))
plt.plot(df["DATE"], df["Diff_Consumption"])
plt.xlabel("Year")
plt.ylabel("Differenced Power Consumption")
plt.title("Differenced Power Consumption")
plt.show()
```
